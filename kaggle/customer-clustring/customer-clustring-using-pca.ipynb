<h1 style="background-color:white; color:#145A32 ; font-size:145%"><br>What is Clustering?<br></h1></center>
‚Äã
‚Äã



**In a business context:**  Clustering algorithm is a technique that assists customer segmentation which is a process of classifying similar customers into the same segment. Clustering algorithm helps to better understand customers, in terms of both static demographics and dynamic behaviors. Customer with comparable characteristics often interact with the business similarly, thus business can benefit from this technique by creating tailored marketing strategy for each segment.  Based on that, customers can be provided with discounts, offers, promo codes etc.
As a simple example. A bank wants to give credit card offers to its customers. Currently, they look at the details of each customer and based on this information, decide which offer should be given to which customer.
Now, the bank can potentially have millions of customers. Does it make sense to look at the details of each customer separately and then make a decision? Certainly not! It is a manual process and will take a huge amount of time.
So what can the bank do? One option is to segment its customers into different groups. For instance, the bank can group the customers based on their income

The bank can now make three different strategies or offers, one for each group. Here, instead of creating different strategies for individual customers, they only have to make 3 strategies. This will reduce the effort as well as the time. Clustering is the process of dividing the entire data into groups (also known as clusters) based on the patterns in the data.

<h1 style="background-color:white; color:#145A32 ; font-size:130%">segmentation <br></h1></center>
‚Äã
‚Äã

it is s an integral part of the development of marketing objectives and strategies, where defining those objectives will generally include either :

- (a) an analysis of how products should be sold or developed, based on an analysis of current customer segments

- (b) the identification of new segments as targets for existing products or the development of new products.
‚Äã
- Customer segmentation and clustering are two techniques used in marketing and data analysis to group customers into similar categories. However, there are some differences between the two.
- **Customer segmentation** is the process of dividing a large and heterogeneous customer base into smaller, more homogenous groups based on specific criteria or characteristics. Segmentation is often done based on demographic information, such as age, gender, or income, as well as behavior or usage patterns, such as purchase history or website activity. The goal of customer segmentation is to create targeted marketing strategies that are more relevant to the needs and preferences of specific customer groups. 

‚Äã
‚Äã
‚Äã
<h1 style="background-color:white; color:#145A32 ; font-size:130%">Clustering <br></h1></center>
‚Äã
‚Äã

- **Clustering**, on the other hand, is a statistical technique used to group data points into clusters based on similarities in their attributes or variables. This can be applied to customer data to group customers based on shared characteristics or behaviors, such as spending patterns or product preferences. The goal of clustering is to identify groups or segments of customers that are similar to each other but distinct from other groups, so that tailored marketing or communication strategies can be developed.
‚Äã
‚Äã
In summary, customer segmentation is a marketing strategy that aims to divide customers into groups based on specific criteria, while clustering is a data analysis technique that groups data points based on similarities in their attributes or variables. Both techniques are useful for understanding customer behavior and developing targeted marketing strategies.

<h1 style="background-color:white; color:darkorange ; font-size:120%"> "cash_advance_trx" <br></h1></center>

- When a customer makes a purchase using the "cash in advance" payment method, it means that they are required to pay for the goods or services upfront before they are delivered. This payment method is often used by businesses that are new or do not have a well-established credit history, as it reduces the risk of payment default. It is also common for online purchases, where the seller may not have the ability to verify the customer's identity or creditworthiness.For example, if you were to purchase a product online using a "cash in advance" payment method, you would typically be required to transfer the funds to the seller's account before they ship the product to you. This ensures that the seller receives payment for the product before it is delivered, reducing the risk of fraud or payment default.The "cash_advance_trx" metric then represents the number of transactions made using this payment method. For a business, this metric can be useful in understanding the payment preferences of their customers and determining whether there is a demand for alternative payment methods. It can also be used to monitor the effectiveness of marketing campaigns aimed at promoting this payment method or to identify potential fraud risks associated with these transactions.

<h1 style="background-color:white; color:darkorange ; font-size:120%">  "purchases_trx" <br></h1></center>

- The "purchases_trx" metric represents the number of purchase transactions made by customers or clients. It is a count of the total number of purchases that have been made during a given time period, such as a day, week, or month.The "Number of purchase transactions made" portion of the metric refers to the total number of transactions made by customers or clients. This could include purchases made online or in-store, by individual customers or businesses. The "discrete (integer)" portion of the metric refers to the fact that the value of the metric is a whole number (i.e., an integer), rather than a decimal or fraction. In other words, it represents a count of the number of transactions made, rather than a monetary value.The "purchases_trx" metric is commonly used by businesses to track the number of transactions made and monitor trends in customer behavior. It can also be used to identify potential areas for improvement, such as increasing customer retention or improving the customer experience. Additionally, businesses can use this metric to calculate other important performance indicators, such as average transaction value, conversion rate, and customer lifetime value.

Specifically, in the context of credit cards, "tenure" usually refers to the number of years that a customer has held an account with a credit card issuer.The length of a customer's tenure can be an important metric for credit card companies, as it can be used to predict customer loyalty and overall satisfaction. Customers who have been with a company for a longer period of time are more likely to be loyal and generate more revenue over time.Additionally, credit card companies may use tenure as a factor when determining credit limits or interest rates for individual customers. Customers with longer tenures and a good payment history are generally seen as less risky and may be eligible for higher credit limits or lower interest rates. Overall, tenure is an important metric for credit card companies to track and can provide valuable insights into customer behavior and preferences.

```python
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False) # Total number of null values
    percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending=False) # Percentage of values that are null
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent']) # Putting the above two together
    return missing_data

missing_check(df)
```

```python
# Categorize customers into 4 groups based on their purchase behavior
# This uses numpy.where() to create nested conditional logic
# Logic flow:
#   1. If both purchases are 0 ‚Üí "None_Of_the_Purchases"
#   2. Else if one-off > 0 AND installment = 0 ‚Üí "One_Of_Purchase"
#   3. Else if one-off = 0 AND installment > 0 ‚Üí "Installment_Purchases"
#   4. Else (both > 0) ‚Üí "Both_the_Purchases"

df3['Purchase_Type'] = np.where(
    (df3['ONEOFF_PURCHASES'] == 0) & (df3['INSTALLMENTS_PURCHASES'] == 0),
    'None_Of_the_Purchases',
    np.where(
        (df3['ONEOFF_PURCHASES'] > 0) & (df3['INSTALLMENTS_PURCHASES'] == 0), 
        'One_Of_Purchase',
        np.where(
            (df_purchases['ONEOFF_PURCHASES'] == 0) & (df_purchases['INSTALLMENTS_PURCHASES'] > 0),
            'Installment_Purchases',
            'Both_the_Purchases'
        )
    )
)

# Display the distribution of customers across purchase types
df3['Purchase_Type'].value_counts()
```

A higher utilization rate indicates presense of credit risk. Hence, a lower utilization rate (balance-to-limit ratio) is desirable

```python
# Feature Engineering: Calculate Credit Utilization Rate (Limit Usage)
# This is the ratio of balance to credit limit - a key metric in credit risk assessment
# Formula: Current Balance / Credit Limit
# 
# Interpretation:
#   - Low ratio (< 0.3): Good - indicates responsible credit usage
#   - Medium ratio (0.3-0.7): Moderate risk
#   - High ratio (> 0.7): High risk - customer is using most of their available credit
#
# This is also known as "utilization rate" in credit scoring

df3['Limit_Usage'] = df3['BALANCE'] / df3['CREDIT_LIMIT']
```

Data Normalization:
- Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to a common scale, without distorting differences in the ranges of values. For machine learning, every dataset does not require normalization. It is required only when features have different ranges.We have data on frequency, which varie from 0 to 1 and data on payments that have a much greater scale. To implement a clustering algorithm it is important to put the data in the same scale, once the distance between the data is taken into account

```python
# Data Standardization (Feature Scaling)
# Apply StandardScaler to transform all numerical features to the same scale
# 
# Why standardization is crucial for clustering:
#   1. Features have different ranges (e.g., frequency: 0-1 vs purchases: 0-50000)
#   2. Distance-based algorithms (K-Means, PCA) are sensitive to scale
#   3. Without scaling, features with larger ranges dominate the clustering
# 
# StandardScaler formula: z = (x - mean) / standard_deviation
#   - Transforms data to have mean=0 and standard deviation=1
#   - Preserves the shape of the distribution
#   - Makes all features equally important in distance calculations

SS = StandardScaler()
x_scaled = pd.DataFrame(SS.fit_transform(df5[x_num]), columns=x_num)
x_scaled.head()
```

**We will use PCA for our dimensionality reduction.**

-To explain briefly what PCA does under the hood is that it finds new dimension/axis for the dataset such that it explains maximum variance. That axis is then the first principal component. Then it chooses another component perpendicular to first principal component which explains maximum variance

```python
# Apply PCA (Principal Component Analysis) for Dimensionality Reduction
# PCA reduces 17 features to fewer components while retaining maximum variance
# 
# What PCA does:
#   1. Finds new axes (principal components) that explain maximum variance
#   2. First component explains most variance, second explains second-most, etc.
#   3. Each component is a linear combination of original features
#   4. Components are orthogonal (perpendicular) to each other
# 
# Benefits:
#   - Reduces computational complexity
#   - Removes multicollinearity
#   - Reduces noise in the data
#   - Makes visualization possible

pca_model = PCA(n_components=17)
X_PCA = pca_model.fit_transform(concat_df)

# Calculate cumulative variance explained by each component
# This helps us decide how many components to keep
pca_var = pca_model.explained_variance_ratio_
np.cumsum(pca_var)
```

```python
# Cumulative Variance explains
var1 = np.cumsum(np.round(pca_model.explained_variance_ratio_, decimals=6)*100)
var1
```

```python
# Summary table showing the Eigen Vectors, Eigen Values and the variance explained by each of the component(eigenvector)

vec_val = pd.DataFrame({'Eigen_Values':pca_model.explained_variance_, 'Cumulative_Variance':var1}, 
                       index=range(1,18)).round(4)
vec_val
```

> Conclusion:
    
> 1.The number of components choosen are 7 that cumulatively explain 85% variation and each individual component explains more than 0.7 variance.

> 2.By considering the 7 dimensions out of the 17 variables we are losing only about 15% of the variation (information) of the data.

```python
# Reduce to 7 Principal Components based on variance analysis
# From the previous cell, we determined that 7 components explain ~85% of variance
# This is a good trade-off: retaining most information while reducing from 17 to 7 dimensions
# 
# Decision criteria:
#   - 7 components capture 85% of total variance
#   - Each component explains > 0.7 eigenvalue (Kaiser criterion)
#   - We lose only 15% of information but gain significant dimensionality reduction

PCA_7 = PCA(n_components=7)
X_PCA_7 = PCA_7.fit_transform(concat_df)

# Create dataframe with principal component columns for easier analysis
PC = pd.DataFrame(X_PCA_7, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
PC
```

```python
# Find the optimal number of clusters using multiple evaluation metrics
# Testing K=3 to K=8 clusters to determine the best segmentation
# 
# Metrics used:
#   1. WCSS (Within-Cluster Sum of Squares / Inertia): Lower is better
#      - Measures compactness of clusters
#      - Sum of squared distances from points to their cluster centroids
#   
#   2. Silhouette Score: Higher is better (range: -1 to 1)
#      - Measures how similar points are to their own cluster vs other clusters
#      - > 0.5: Good clustering
#      - 0.3-0.5: Reasonable clustering
#      - < 0.3: Poor clustering
# 
# K-Means++ initialization: Smart centroid placement for faster convergence

wcss = []
sil_kmeans = []

for i in range(3, 9):
    # K-Means Clustering with k-means++ initialization
    kmeans = KMeans(n_clusters=i, n_init=100, init='k-means++', random_state=0)
    kmeans.fit(X_PCA_7)
    
    # Inertia (WCSS) for Elbow Method
    in_km = kmeans.inertia_
    wcss.append(in_km)
    
    # Silhouette Score for cluster quality assessment
    sil_km = silhouette_score(X_PCA_7, kmeans.labels_)
    sil_kmeans.append(sil_km)
    
    print('Number of clusters:', i)
    print('KMeans Inertia (WCSS):', round(in_km, 2))
    print('Silhouette Score for KMeans:', round(sil_km, 4))
    print()
```

```python
# Additional Clustering Evaluation Metrics
# Adding Calinski-Harabasz Score and Davies-Bouldin Index for comprehensive evaluation
# 
# Calinski-Harabasz Score (Variance Ratio Criterion):
#   - Measures ratio of between-cluster variance to within-cluster variance
#   - Higher is better (better separated, more compact clusters)
#   - No fixed range, compare relative values
# 
# Davies-Bouldin Index:
#   - Measures average similarity between each cluster and its most similar cluster
#   - Lower is better (0 is perfect, higher means more overlap)
#   - Range: 0 to infinity
# 
# Combined with Silhouette Score and Elbow Method, these provide a complete picture

from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score

ch_scores = []
db_scores = []

print("Comprehensive Clustering Evaluation")
print("=" * 60)

for k in range(3, 9):
    kmeans = KMeans(n_clusters=k, n_init=100, init='k-means++', random_state=0)
    labels = kmeans.fit_predict(X_PCA_7)
    
    # Calculate all metrics
    ch_score = calinski_harabasz_score(X_PCA_7, labels)
    db_score = davies_bouldin_score(X_PCA_7, labels)
    sil_score = silhouette_score(X_PCA_7, labels)
    
    ch_scores.append(ch_score)
    db_scores.append(db_score)
    
    print(f'\nK = {k}:')
    print(f'  Calinski-Harabasz Score: {ch_score:.2f} (higher is better)')
    print(f'  Davies-Bouldin Index: {db_score:.4f} (lower is better)')
    print(f'  Silhouette Score: {sil_score:.4f} (higher is better)')

print("\n" + "=" * 60)
print("\nBest K based on each metric:")
print(f"  Best by Calinski-Harabasz: K={ch_scores.index(max(ch_scores)) + 3}")
print(f"  Best by Davies-Bouldin: K={db_scores.index(min(db_scores)) + 3}")
print(f"  Best by Silhouette: K={sil_kmeans.index(max(sil_kmeans)) + 3}")
```

# <p style="background-color:#145A32;font-family:newtimeroman;font-size:110%;color:orange;text-align:center;border-radius:15px 50px; padding:7px">Elbow Method 
</p>

>- When clustering a dataset, we want to group similar data points together into clusters. However, we need to determine how many clusters we should create. The elbow method is a technique to help us determine the optimal number of clusters.
To use the elbow method, we first calculate the within-cluster sum of squares (WCSS) for different numbers of clusters. The WCSS measures the sum of the squared distances between each data point and the centroid of its assigned cluster. In other words, it measures how tightly the data points are clustered around their respective centroids.
We then plot the WCSS values against the number of clusters, creating an "elbow curve". The curve will typically decrease as the number of clusters increases, because increasing the number of clusters will tend to reduce the WCSS. However, as we add more and more clusters, the improvement in WCSS will start to level off. The point where the curve starts to flatten out is referred to as the "elbow" point.
The optimal number of clusters can be chosen as the point just before the elbow, where the WCSS improvement begins to level off. This is a trade-off between having enough clusters to capture the important patterns in the data, but not so many that we overfit to the noise in the data.
The elbow method is a simple and effective technique to help us choose the number of clusters when performing clustering analysis. However, it's worth noting that it's not always easy to identify a clear elbow point, and sometimes it may be necessary to use additional techniques to determine the optimal number of clusters.

```python
# Create the final profiling output table with segment sizes, percentages, and feature averages
# This is the complete summary table for business stakeholders
# 
# Table structure:
#   Row 1: Segment sizes (number of customers in each segment)
#   Row 2: Segment percentages (% distribution of customers)
#   Remaining rows: Average feature values for each segment
# 
# This table answers key business questions:
#   - How many customers are in each segment?
#   - What percentage of our customer base does each segment represent?
#   - What are the spending patterns and behaviors of each segment?
#   - Which segments are most valuable (high purchases, low risk)?
# 
# Use this table to develop targeted marketing strategies for each segment

Profiling_output_final = pd.concat([Seg_size.T, Seg_Pct.T, Profiling_output], axis=0)

# Add descriptive column names for each clustering solution
Profiling_output_final.columns = [
    'Overall',
    'KM3_1', 'KM3_2', 'KM3_3',
    'KM4_1', 'KM4_2', 'KM4_3', 'KM4_4',
    'KM5_1', 'KM5_2', 'KM5_3', 'KM5_4', 'KM5_5',
    'KM6_1', 'KM6_2', 'KM6_3', 'KM6_4', 'KM6_5', 'KM6_6',
    'KM7_1', 'KM7_2', 'KM7_3', 'KM7_4', 'KM7_5', 'KM7_6', 'KM7_7',
    'KM8_1', 'KM8_2', 'KM8_3', 'KM8_4', 'KM8_5', 'KM8_6', 'KM8_7', 'KM8_8'
]

Profiling_output_final
```
